```python
# Dependencies
import pandas as pd
```

```python
# List of file names to be combined 
data_2023 = ['Resources\JC-202309-citibike-tripdata.csv', 'Resources\JC-202310-citibike-tripdata.csv', 'Resources\JC-202311-citibike-tripdata.csv']

# Create an empty list to store each dataframe
df23_list = []

# Loop through each file in the list
for file in data_2023:
    # Read the file into a dataframe
    df23 = pd.read_csv(file)

    # Append the dataframe to the list
    df23_list.append(df23)

# Concatenate all dataframes in the list into one dataframe
df23 = pd.concat(df23_list)

# Show dataframe
df23
```

```python
# print column headers
column_headers = df23.columns.values.tolist()
print("Column Headers:", column_headers)
```

```python
# Save the dataframe to a csv 
df23.to_csv('Resources\citibike_last3months.csv', index=False)
```
